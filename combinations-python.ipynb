```python
import itertools
import matplotlib.pyplot as plt
from collections import Counter
import random

# Alfabeto sin la letra "L"
alfabeto_sin_L = [
    'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 
    'M', 'N', 'O', 'P', 'Q', 'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y', 'Z'
]

numeros_aleatorios = list(range(1, 26))
random.shuffle(numeros_aleatorios)
pesos = {letra: peso for letra, peso in zip(alfabeto_sin_L, numeros_aleatorios)}
print(pesos)

# Función para convertir combinación de letras a un código de 3 dígitos
def convertir_combinacion_a_entero(letras):
    letras_numero = 0
    for i in range(5):
        posicion = alfabeto_sin_L.index(letras[i])
        peso = pesos[letras[i]]  # Usamos el peso en lugar de solo la posición
        letras_numero = (letras_numero * 25 + posicion + peso) % 1000
    return letras_numero

# Generar todas las combinaciones posibles de 5 letras sin repetición
combinaciones = itertools.permutations(alfabeto_sin_L, 5)

# Convertir cada combinación a un código y contar las colisiones
codigos = [convertir_combinacion_a_entero(comb) for comb in combinaciones]
contador_codigos = Counter(codigos)

# Graficar los resultados
codigos, frecuencias = zip(*contador_codigos.items())

plt.bar(codigos, frecuencias, edgecolor='black')
plt.xlabel('Código de 3 dígitos')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de cada código de 3 dígitos')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
min_frecuencia = min(contador_codigos.values())
codigos_menos_probables = [codigo for codigo, frecuencia in contador_codigos.items() if frecuencia == min_frecuencia]

# Encontrar las combinaciones de letras que generan estos códigos
combinaciones_menos_probables = [
    comb for comb in itertools.permutations(alfabeto_sin_L, 5) 
    if convertir_combinacion_a_entero(comb) in codigos_menos_probables
]

print(f"La(s) combinación(es) menos probable(s) tiene(n) una frecuencia de {min_frecuencia} y son:")
for comb in combinaciones_menos_probables:
    print(''.join(comb), '->', convertir_combinacion_a_entero(comb))
```
